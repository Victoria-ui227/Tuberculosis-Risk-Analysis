# TUBERCULOSIS RISK ANALYSIS IN WHO REGIONS.


## OVERVIEW 

Tuberculosis (TB) is an infectious disease caused by bacteria that most often affects the lungs. It spreads through the air when people with TB cough, sneeze or spit. It is preventable and curable.

Approximately one-quarter of the world's population carries the TB bacteria, with only 5-10% of those infected developing active TB disease. Infected individuals without symptoms cannot transmit the disease. TB disease is typically treated with antibiotics and can be fatal if left untreated. 
In some countries, the Bacille Calmette-Gu√©rin (BCG) vaccine is administered to infants and young children, effectively preventing TB-related deaths and safeguarding them from severe forms of the disease.


## PROBLEM STATEMENT

TB is one of the major leading causes of death in the world from a single infectious agent, following three years in which it was replaced by coronavirus disease (COVID-19).

It was also the leading killer of people with HIV and a major cause of deaths related to antimicrobial resistance. It is present in all countries and age groups. It is therefore a disease of public health importance and there have been efforts made to combat the disease globally which has saved an estimated 79 million lives since the year 2000.

This project aims to analyze the burden of TB in high risk WHO regions in terms of incidence and mortality rates.Visuase the health systems performance in response to TB and finally build a logistic model to predict case detection rate for the disease. 
The insights drawn from the analysis would help inform public health decision making in policies aimed at preventing TB in populations.  


## DATA UNDERSTANDING 

The datasets used in the project are sourced from World Health Organization (WHO) https://www.who.int/data/gho/data/themes/tuberculosis.

They include three csv files: MDR_RR_TB_burden_estimates_2025-06-06, TB_burden_countries_2025-06-07 and TB_data_dictionary_2025-06-06. 

The files contains data reported by WHO and estimates of tuberculosis burden generated by WHO for the Global tuberculosis report.The report provides a comprehensive and up-to-date assessment of the TB epidemic, and of progress in prevention, diagnosis and treatment of the disease at global, regional and country levels.

The data dictionary provides description of the data variables in the other two csv files which is also provided by the text description file.

The MDR_RR_TB_burden_estimates_2025-06-06 file contains data on MDR-TB which is a a form of TB caused by bacteria that do not respond to isoniazid and rifampicin, the two most effective first-line TB drugs.

The TB_burden_countries_2025-06-07 file captures the overall burden estimates of disease in the WHO regions.

### Key Predictive Features for TB Risk:

##### 1. Direct incidence metrics(targets)
We use the variables that represent incidence in the different regions:

e_inc_100k - Estimated TB incidence per 100k population

e_inc_num - Absolute number of new TB cases and 

e_inc_tbhiv_100k - TB-HIV coinfection rate (HIV increases TB risk).

##### 2. Mortality and Severity Indicators
e_mort_100k: TB mortality rate (high mortality ‚âà poor healthcare access).

e_mort_tbhiv_num: Deaths in HIV+ TB patients (indicates comorbid severity).

cfr_pct: Case fatality ratio (untreated TB burden).

##### 3. Drug Resistance (Critical for Outbreak Risk)
e_inc_rr_num: Rifampicin-resistant TB cases (harder to treat).

e_rr_pct_new: % of new TB cases with drug resistance.

##### 4. Healthcare System Performance
c_cdr: Case detection rate (low ‚âà underdiagnosis).

e_rr_in_notified_labconf_pulm: Lab-confirmed resistant TB (data quality signal).

##### 5. HIV Coinfection (Major Risk Multiplier)
e_tbhiv_prct: % of TB cases with HIV (strong risk factor).


We'll subset the data to remain with relevant columns

```python
TB1= TB_estimates1[['country','iso2','iso3','iso_numeric','g_whoregion','year','e_pop_num','e_inc_100k','e_inc_num','e_inc_tbhiv_100k',
         'e_mort_100k','e_mort_tbhiv_num','cfr_pct','c_cdr','e_tbhiv_prct','e_mort_tbhiv_100k'
         ]]

TB1
```

For the drug resistance data

```python
drug_resistance= TB_est2[['country','e_inc_rr_num', 'e_rr_pct_new', 'e_rr_in_notified_labconf_pulm','e_rr_pct_ret']]

drug_resistance
```

```python
TBdf= pd.merge(TB1, drug_resistance, on='country', how='inner') 
TBdf
```

```python
TBdf.info()
```

### 1. Incidence prediction

##### Feature engineering

```python
# Add a high-risk column (1 if high TB incidence, else 0)
TBdf['high_risk'] = TBdf['e_inc_100k'].apply(lambda x: 1 if x >= 100 else 0)
TBdf.head()
```

```python
TBdf['high_risk'].value_counts()
```

```python
TBdf['untreated_cases'] = TBdf['e_inc_num'] * (1 - TBdf['c_cdr']/100)  # Estimated undiagnosed cases
TBdf['hiv_impact'] = TBdf['e_tbhiv_prct'] * TBdf['e_inc_num']          # HIV-driven case burden
```

```python
#features
features=[
    'untreated_cases',
    'hiv_impact',
    'e_tbhiv_prct', #HIV burden 
    'c_cdr',        #Healthcare access
    'cfr_pct'       #mortality proxy
]
```

The model predicts whether a region is high-risk (1) or low-risk (0) for TB based on:

Features: Untreated cases, HIV impact, HIV burden (e_tbhiv_prct), healthcare access (c_cdr), and mortality proxy (cfr_pct).

Target: Binary high_risk column (1 = high risk, 0 = low risk).

```python
imputer_median = SimpleImputer(strategy='median')
TBdf[['untreated_cases', 'hiv_impact','e_tbhiv_prct','c_cdr','cfr_pct']] = imputer_median.fit_transform(TBdf[['untreated_cases', 'hiv_impact','e_tbhiv_prct','c_cdr','cfr_pct']])
```

```python
scaler = StandardScaler()
X_scaled = scaler.fit_transform(TBdf[features])
```

```python
#define X and y variables 
X= X_scaled
y= TBdf['high_risk']

#traintest split the data 
X_train, X_test, y_train, y_test = train_test_split(
    X_scaled, 
    y, 
    test_size=0.2, 
    random_state=42
)

#Train the logistic regresssion model
model = LogisticRegression(class_weight='balanced', C=1.0,max_iter=1000)  # gradient descent to control for class imbalance
model.fit(X_train, y_train)
```

#### Evaluate model performance

```python
#Calculating Log Loss 

# Predict probabilities for the training set
y_train_pred_proba = model.predict_proba(X_train)
train_log_loss = log_loss(y_train, y_train_pred_proba)
print(f"Training Log Loss: {train_log_loss:.4f}")

# Predict probabilities for the test set
y_pred_proba = model.predict_proba(X_test)
test_log_loss = log_loss(y_test, y_pred_proba)
print(f"Test Log Loss: {test_log_loss:.4f}")
```

```python
#Plotting Log Loss

labels = ['Training Log Loss', 'Test Log Loss']
values = [train_log_loss, test_log_loss]

plt.figure(figsize=(8, 6))
plt.bar(labels, values, color=['skyblue', 'lightcoral'])
plt.ylabel('Log Loss')
plt.title('Logistic Regression Model Log Loss')
plt.ylim(0, max(values) * 1.2) # Set y-axis limit slightly above max value for better visualization
plt.grid(axis='y', linestyle='--', alpha=0.7)

# Add text labels on top of the bars
for i, val in enumerate(values):
    plt.text(i, val + 0.01, f'{val:.4f}', ha='center', va='bottom')

plt.show()
```

Both the training and test log loss values (0.1730 and 0.1812, respectively) are relatively low.The test log loss (0.1812) is slightly higher than the training log loss (0.1730).

The fact that the test log loss is only marginally higher than the training log loss suggests that the model is not significantly overfitting the training data

```python
y_pred = model.predict(X_test)
print(classification_report(y_test, y_pred))
print("ROC-AUC:", roc_auc_score(y_test, y_pred_proba[:, 1]))
```

#### Interpretation

The model achieves an overall accuracy of 85% and a strong ROC-AUC of 0.87, indicating excellent ability to distinguish between high-risk and not-high-risk cases.

Class 0 (Not High Risk) is predicted more accurately, with high precision (0.88) and recall (0.90).

Class 1 (High Risk) has slightly lower performance (precision = 0.79, recall = 0.75), suggesting the model misses some true high-risk cases.

F1-scores reflect this difference: 0.89 for class 0, and 0.77 for class 1.

The macro average (0.83) shows good balance across classes, and the weighted average (0.85) accounts for class imbalance.

#### Feature Implications:
The features (untreated_cases, hiv_impact, etc.) appear to be good predictors since:

The AUC (0.87) suggests strong feature-target relationships

The model achieves good precision/recall without severe class bias

#### Potential Improvements:
If identifying high-risk cases (class 1) is critical, we might want to improve recall (75% means 25% of high-risk cases are missed).

Check for feature engineering opportunities (e.g., interactions between hiv_impact and e_tbhiv_prct).

### 2. Mortality rates in different regions.

#### Filter and Visualize Countries with High TB mortality

```python
# Filter for countries with mortality > 100 for both TB and TB-HIV
high_mortality_df = TBdf.loc[
    (TBdf['e_mort_100k'] > 100) & (TBdf['e_mort_tbhiv_100k'] > 100)
]
```

```python
plt.figure(figsize=(12, 6))

# --- Plot 1: TB Mortality ---
sns.lineplot(
    data=high_mortality_df,
    x='year',
    y='e_mort_100k',
    hue='country',
    marker='o',
    legend='full'  
)
plt.title('TB Mortality Rate per 100,000 Population (2000-2023)')
plt.xlabel('Year')
plt.ylabel('Mortality Rate')
plt.xlim(2000, 2023)  
plt.xticks(range(2000, 2024, 5))  # Show ticks every 5 years for clarity
plt.tight_layout()
plt.show()

# --- Plot 2: TB-HIV Mortality ---
plt.figure(figsize=(12, 6))
sns.lineplot(
    data=high_mortality_df,
    x='year',
    y='e_mort_tbhiv_100k',
    hue='country',
    marker='o',
    legend='full'  
)
plt.title('TB-HIV Mortality Rate per 100,000 Population (2000-2023)')
plt.xlabel('Year')
plt.ylabel('Mortality Rate')
plt.xlim(2000, 2023)  
plt.xticks(range(2000, 2024, 5))  # Show ticks every 5 years
plt.tight_layout()
plt.show()
```

From the plots, both show broadly similar trends, with many countries experiencing a peak in mortality rates in the early to mid-2000s, followed by a general decline. This suggests that the decline in overall TB mortality is significantly influenced by the reduction in TB-HIV co-mortality.

The significant drops observed in overall TB mortality in these countries from the mid-2000s onwards are mirrored by similar sharp declines in TB-HIV mortality, likely due to the scale-up of antiretroviral therapy (ART) and TB/HIV collaborative activities. This suggests that effective HIV treatment and prevention directly contribute to reducing TB mortality.

From this, we conclude that healthcare system performance has significantly improved since 2010 and global public health efforts to prevent TB has shown tremendous success.
